```python
import pandas as pd
```

```python
signin = pd.read_csv('/Users/olgabot/Downloads/cupcakes_and_coding_2018 - sign-in.csv')
print(signin.shape)
signin.head()
```

```python
signin.tail()
```

```python
signin = signin.dropna(subset=['First Name'])
signin.tail()
```

```python
signin['First Name'] = signin['First Name'].str.capitalize()
signin['Last Name'] = signin['Last Name'].str.capitalize()
```

```python
signin['Name'] = signin.apply(lambda x: '{First Name} {Last Name}'.format(**x), axis=1)
signin.head()
```

```python
sorted_names = sorted(signin.Name.unique())
sorted_names
```
